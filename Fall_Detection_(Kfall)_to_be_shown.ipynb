```python
"""
Fall Detection System using CNN-LSTM Hybrid Model
Dataset: KFall Dataset from Kaggle
Description: Detects fall vs non-fall activities using inertial sensor data
Author: ML Project Implementation
"""

# ==================== INSTALLATION AND IMPORTS ====================
!pip install kagglehub -q
!pip install tensorflow scikit-learn matplotlib seaborn pandas numpy openpyxl -q

import kagglehub
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (confusion_matrix, classification_report,
                              accuracy_score, precision_score, recall_score,
                              f1_score, roc_curve, auc)
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import (Dense, Dropout, LSTM, Conv1D, MaxPooling1D,
                                      Flatten, Input, BatchNormalization, Bidirectional)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("="*80)
print("FALL DETECTION SYSTEM - KFall Dataset")
print("Using CNN-LSTM Hybrid Architecture")
print("="*80)

# ==================== DATASET DOWNLOAD ====================
print("\n[1/9] Downloading KFall Dataset...")
path = kagglehub.dataset_download("usmanabbasi2002/kfall-dataset")
print(f"✓ Dataset downloaded to: {path}")

# ==================== DATASET EXPLORATION ====================
print("\n[2/9] Exploring Dataset Structure...")

# Find CSV files and label files
csv_files = glob.glob(os.path.join(path, "**/*.csv"), recursive=True)
xlsx_files = glob.glob(os.path.join(path, "**/*.xlsx"), recursive=True)

print(f"✓ Found {len(csv_files)} CSV sensor data files")
print(f"✓ Found {len(xlsx_files)} Excel label files")

# Display directory structure
print("\nDataset Structure:")
for root, dirs, files in os.walk(path):
    level = root.replace(path, '').count(os.sep)
    indent = ' ' * 2 * level
    print(f'{indent}{os.path.basename(root)}/')
    if level < 2:
        subindent = ' ' * 2 * (level + 1)
        for file in files[:3]:
            print(f'{subindent}{file}')
        if len(files) > 3:
            print(f'{subindent}... and {len(files) - 3} more files')

# ==================== DATA LOADING AND LABELING ====================
print("\n[3/9] Loading and Labeling Data...")

# KFall Dataset Structure:
# - Task codes D01-D21: Activities of Daily Living (ADL) - Non-Fall
# - Task codes F01-F15: Fall activities - Fall
# Task code format: SAxTyyRzz (e.g., SA01T01R01)
# SA: Subject ID, T: Task, R: Repetition

def extract_task_code(filename):
    """Extract task code from filename"""
    basename = os.path.basename(filename).replace('.csv', '')
    # Format: SAxTyyRzz -> extract Tyy part
    if 'T' in basename and 'R' in basename:
        start_idx = basename.index('T')
        end_idx = basename.index('R')
        task_code = basename[start_idx:end_idx]
        return task_code
    return None

def is_fall_activity(task_code):
    """Determine if task code represents a fall activity"""
    if task_code is None:
        return False
    # Extract task number (e.g., T20 -> 20)
    task_num = int(task_code.replace('T', ''))
    # Tasks 1-21 are ADL (Non-Fall), Tasks 20-34 are Falls
    # Based on KFall dataset: F01-F15 corresponds to higher task numbers
    return task_num > 21

# Load all sensor data
all_data = []
all_labels = []
file_info = []

print("\nLoading sensor data files...")
for i, csv_file in enumerate(csv_files):
    try:
        # Read CSV file
        df = pd.read_csv(csv_file)

        # Extract task code
        task_code = extract_task_code(csv_file)

        # Determine label based on task code or filename
        filename = os.path.basename(csv_file).lower()

        # Label determination logic
        if task_code and is_fall_activity(task_code):
            label = 1  # Fall
        elif 'fall' in filename or any(f'f{i:02d}' in filename for i in range(1, 16)):
            label = 1  # Fall
        else:
            label = 0  # Non-Fall (ADL)

        # Select sensor columns (acceleration, gyroscope, euler angles)
        # KFall format: TimeStamp, FrameCounter, Acc_X/Y/Z, Gyr_X/Y/Z, Euler_X/Y/Z
        sensor_cols = [col for col in df.columns if any(x in col.lower()
                       for x in ['acc', 'gyr', 'euler', 'angular', 'angle'])]

        if len(sensor_cols) > 0:
            sensor_data = df[sensor_cols].values
            all_data.append(sensor_data)
            all_labels.append(label)
            file_info.append({
                'file': os.path.basename(csv_file),
                'task': task_code,
                'label': 'Fall' if label == 1 else 'ADL',
                'length': len(sensor_data)
            })

        if (i + 1) % 500 == 0:
            print(f"  Processed {i + 1}/{len(csv_files)} files...")

    except Exception as e:
        print(f"  Error loading {csv_file}: {e}")

print(f"\n✓ Successfully loaded {len(all_data)} activity samples")

# Display sample information
sample_df = pd.DataFrame(file_info[:10])
print("\nSample Files (first 10):")
print(sample_df.to_string(index=False))

# Label distribution
labels_array = np.array(all_labels)
unique, counts = np.unique(labels_array, return_counts=True)
print(f"\n✓ Label Distribution:")
print(f"  ADL (Non-Fall): {counts[0]} samples ({counts[0]/len(labels_array)*100:.1f}%)")
print(f"  Fall: {counts[1]} samples ({counts[1]/len(labels_array)*100:.1f}%)")

# Display sample sensor data
print(f"\n✓ Sample Sensor Data Shape: {all_data[0].shape}")
print(f"  Sequence Length: {all_data[0].shape[0]} frames")
print(f"  Number of Features: {all_data[0].shape[1]} (sensor channels)")

# ==================== SEQUENCE CREATION ====================
print("\n[4/9] Creating Fixed-Length Sequences...")

def create_sequences(data_list, labels, sequence_length=200):
    """
    Create fixed-length sequences from variable-length sensor data
    """
    X_sequences = []
    y_labels = []

    for data, label in zip(data_list, labels):
        # Handle variable sequence lengths
        if len(data) >= sequence_length:
            # If longer, take multiple overlapping windows
            num_windows = (len(data) - sequence_length) // 50 + 1
            for i in range(num_windows):
                start_idx = i * 50
                end_idx = start_idx + sequence_length
                if end_idx <= len(data):
                    X_sequences.append(data[start_idx:end_idx])
                    y_labels.append(label)
        else:
            # If shorter, pad with zeros
            padded = np.zeros((sequence_length, data.shape[1]))
            padded[:len(data)] = data
            X_sequences.append(padded)
            y_labels.append(label)

    return np.array(X_sequences), np.array(y_labels)

# Create sequences
sequence_length = 200  # 2 seconds at 100 Hz
X, y = create_sequences(all_data, all_labels, sequence_length)

print(f"✓ Created sequences:")
print(f"  X shape: {X.shape} (samples, timesteps, features)")
print(f"  y shape: {y.shape}")

# Convert labels to categorical
y_categorical = to_categorical(y, num_classes=2)

# ==================== DATA NORMALIZATION ====================
print("\n[5/9] Normalizing Data...")

# Reshape for normalization
n_samples, n_timesteps, n_features = X.shape
X_reshaped = X.reshape(-1, n_features)

# Normalize
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X_reshaped)
X = X_normalized.reshape(n_samples, n_timesteps, n_features)

print(f"✓ Data normalized using StandardScaler")

# ==================== TRAIN-VAL-TEST SPLIT ====================
print("\n[6/9] Splitting Data (Train: 60%, Val: 20%, Test: 20%)...")

# First split: separate test set (20%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y_categorical, test_size=0.2, random_state=42, stratify=y
)

# Second split: separate train and validation (75%-25% of remaining)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=42,
    stratify=np.argmax(y_temp, axis=1)
)

print(f"✓ Training set: {X_train.shape}")
print(f"✓ Validation set: {X_val.shape}")
print(f"✓ Test set: {X_test.shape}")

# ==================== MODEL BUILDING ====================
print("\n[7/9] Building CNN-LSTM Hybrid Model...")

def build_fall_detection_model(input_shape):
    """
    Build hybrid CNN-LSTM model for fall detection
    Architecture: CNN for feature extraction + LSTM for temporal modeling
    """
    model = Sequential([
        # CNN layers for spatial feature extraction
        Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=input_shape),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.3),

        Conv1D(filters=128, kernel_size=5, activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.3),

        Conv1D(filters=256, kernel_size=3, activation='relu'),
        BatchNormalization(),
        MaxPooling1D(pool_size=2),
        Dropout(0.3),

        # Bidirectional LSTM for temporal modeling
        Bidirectional(LSTM(128, return_sequences=True)),
        Dropout(0.4),
        Bidirectional(LSTM(64, return_sequences=False)),
        Dropout(0.4),

        # Dense layers for classification
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(2, activation='softmax')  # Binary classification: ADL vs Fall
    ])

    return model

# Build model
input_shape = (X_train.shape[1], X_train.shape[2])
model = build_fall_detection_model(input_shape)

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy',
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall')]
)

print("\n✓ Model Architecture:")
model.summary()

# ==================== MODEL TRAINING ====================
print("\n[8/9] Training Model...")

# Callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_fall_detection_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# Train model
print("\nTraining in progress...")
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

print("\n✓ Training completed!")

# ==================== MODEL EVALUATION ====================
print("\n[9/9] Evaluating Model on Test Set...")

# Load best model
model.load_weights('best_fall_detection_model.h5')

# Make predictions
y_pred_proba = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)
y_true = np.argmax(y_test, axis=1)

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='binary')
recall = recall_score(y_true, y_pred, average='binary')
f1 = f1_score(y_true, y_pred, average='binary')
specificity = recall_score(y_true, y_pred, pos_label=0)

# Display results
print("\n" + "="*80)
print("PERFORMANCE METRICS")
print("="*80)
print(f"\n{'Metric':<20} {'Value':<10}")
print("-" * 30)
print(f"{'Accuracy':<20} {accuracy:.4f}")
print(f"{'Precision':<20} {precision:.4f}")
print(f"{'Recall (Sensitivity)':<20} {recall:.4f}")
print(f"{'Specificity':<20} {specificity:.4f}")
print(f"{'F1-Score':<20} {f1:.4f}")

# Classification Report
print("\n" + "="*80)
print("CLASSIFICATION REPORT")
print("="*80)
target_names = ['ADL (Non-Fall)', 'Fall']
print(classification_report(y_true, y_pred, target_names=target_names, digits=4))

# ==================== CONFUSION MATRIX ====================
print("\nGenerating Confusion Matrix...")

cm = confusion_matrix(y_true, y_pred)

# Create figure with better styling
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names,
            yticklabels=target_names,
            cbar_kws={'label': 'Count'},
            annot_kws={'size': 16, 'weight': 'bold'})

plt.title('Confusion Matrix - Fall Detection System',
          fontsize=16, fontweight='bold', pad=20)
plt.ylabel('True Label', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=14, fontweight='bold')

# Add percentage annotations
for i in range(len(cm)):
    for j in range(len(cm)):
        percentage = cm[i, j] / cm[i].sum() * 100
        plt.text(j + 0.5, i + 0.7, f'({percentage:.1f}%)',
                ha='center', va='center', fontsize=12, color='red')

plt.tight_layout()
plt.show()

# Display confusion matrix values
print("\nConfusion Matrix:")
print(f"  True Negatives (ADL → ADL):  {cm[0, 0]}")
print(f"  False Positives (ADL → Fall): {cm[0, 1]}")
print(f"  False Negatives (Fall → ADL): {cm[1, 0]}")
print(f"  True Positives (Fall → Fall):  {cm[1, 1]}")

# ==================== TRAINING HISTORY VISUALIZATION ====================
print("\nVisualizing Training History...")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Accuracy plot
axes[0, 0].plot(history.history['accuracy'], label='Train', linewidth=2, marker='o', markersize=4)
axes[0, 0].plot(history.history['val_accuracy'], label='Validation', linewidth=2, marker='s', markersize=4)
axes[0, 0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Epoch', fontsize=12)
axes[0, 0].set_ylabel('Accuracy', fontsize=12)
axes[0, 0].legend(loc='lower right', fontsize=11)
axes[0, 0].grid(True, alpha=0.3)

# Loss plot
axes[0, 1].plot(history.history['loss'], label='Train', linewidth=2, marker='o', markersize=4)
axes[0, 1].plot(history.history['val_loss'], label='Validation', linewidth=2, marker='s', markersize=4)
axes[0, 1].set_title('Model Loss', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Epoch', fontsize=12)
axes[0, 1].set_ylabel('Loss', fontsize=12)
axes[0, 1].legend(loc='upper right', fontsize=11)
axes[0, 1].grid(True, alpha=0.3)

# Precision plot
axes[1, 0].plot(history.history['precision'], label='Train', linewidth=2, marker='o', markersize=4)
axes[1, 0].plot(history.history['val_precision'], label='Validation', linewidth=2, marker='s', markersize=4)
axes[1, 0].set_title('Model Precision', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Epoch', fontsize=12)
axes[1, 0].set_ylabel('Precision', fontsize=12)
axes[1, 0].legend(loc='lower right', fontsize=11)
axes[1, 0].grid(True, alpha=0.3)

# Recall plot
axes[1, 1].plot(history.history['recall'], label='Train', linewidth=2, marker='o', markersize=4)
axes[1, 1].plot(history.history['val_recall'], label='Validation', linewidth=2, marker='s', markersize=4)
axes[1, 1].set_title('Model Recall', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Epoch', fontsize=12)
axes[1, 1].set_ylabel('Recall', fontsize=12)
axes[1, 1].legend(loc='lower right', fontsize=11)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ==================== ROC CURVE ====================
print("\nGenerating ROC Curve...")

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_pred_proba[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8 , 4))
plt.plot(fpr, tpr, color='darkorange', lw=3, label=f'ROC Curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14, fontweight='bold')
plt.ylabel('True Positive Rate', fontsize=14, fontweight='bold')
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16, fontweight='bold')
plt.legend(loc="lower right", fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\n✓ ROC AUC Score: {roc_auc:.4f}")

# ==================== FINAL SUMMARY ====================
print("\n" + "="*80)
print("FALL DETECTION SYSTEM - FINAL SUMMARY")
print("="*80)
print(f"\n✓ Total Samples Processed: {len(all_data)}")
print(f"✓ Training Samples: {len(X_train)}")
print(f"✓ Validation Samples: {len(X_val)}")
print(f"✓ Test Samples: {len(X_test)}")
print(f"\n✓ Model Performance:")
print(f"  - Accuracy: {accuracy*100:.2f}%")
print(f"  - Sensitivity (Fall Detection Rate): {recall*100:.2f}%")
print(f"  - Specificity (ADL Detection Rate): {specificity*100:.2f}%")
print(f"  - F1-Score: {f1:.4f}")
print(f"  - ROC AUC: {roc_auc:.4f}")
print(f"\n✓ Model saved as: best_fall_detection_model.h5")
print("\n" + "="*80)
print("PROJECT COMPLETED SUCCESSFULLY!")
print("="*80)
```
